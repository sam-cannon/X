# Concatenating Send a Smile and Verbatim Feedback
- all this notebook is is me putting these data frames together to use in RapidNLP or clustering, thats it

```python
import pandas as pd
```

```python
df = pd.read_csv('sas_teams_since_Jan_2020.csv')
```

```python
df.shape
```

```python
df['Feedback_Type'].unique()
```

```python
df.head()
```

```python
df2 = pd.read_csv('teams_verbatims_since_Jan_2020.csv')
```

```python
df2.shape
```

```python
df2 = df2[['Feedback_DateTime', 'Feedback_Verbatim', 'Feedback_SurveyName', 
           'Feedback_Platform', 'Feedback_App', 'Feedback_UserType', 'User_Commercial_HasEducation']]
```

```python
df2.head()
```

```python
df = df[['Feedback_DateTime', 'Feedback_Verbatim', 'Feedback_SurveyName', 
           'Feedback_Platform', 'Feedback_App', 'Feedback_UserType']]
```

```python
df.shape
```

```python
sampling_df = pd.concat([df, df2], axis = 0)
```

```python
sampling_df.isna().sum()
```

```python
sample_df = sampling_df.sample(frac=0.2, replace=False, random_state=1)
```

```python
sample_df.shape
```

```python
sample_df.sort_values('Feedback_DateTime').head()
```

```python
sample_df.to_csv('teams_sas_verbatims_since_Jan_2020_sample.csv', index = False)
```

```python
sampling_df.to_csv('all_teams_sas_verbatims.csv', index = False)
```

## Putting together web dfs here

```python
web_verbs = pd.read_csv('teams_web_verbatims_since_Jan_2020.csv')
web_sas = pd.read_csv('sas_teams_web_since_Jan_2020.csv')
```

```python
web_verbs.head()
```

```python
web_sas.head()
```

```python
verbs = web_verbs[['Feedback_DateTime', 'Feedback_Verbatim', 'Feedback_SurveyName', 
           'Feedback_Platform', 'Feedback_App', 'Feedback_UserType', 'User_Commercial_HasEducation']]
```

```python
sas = web_sas[['Feedback_DateTime', 'Feedback_Verbatim', 'Feedback_SurveyName', 
           'Feedback_Platform', 'Feedback_App', 'Feedback_UserType']]
```

```python
combined = pd.concat([verbs, sas], axis = 0)
```

```python
combined.shape
```

```python
combined.head()
```

```python
combined.isna().sum()
```

```python
combined.to_csv('all_teams_web_sas_and_verbs_since_Jan_2020.csv', index = False)
```
